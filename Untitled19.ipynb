```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
import matplotlib.pyplot as plt

# Try xgboost; if not installed in this Colab, fall back to RandomForest
try:
    from xgboost import XGBClassifier
    HAS_XGB = True
except ImportError:
    HAS_XGB = False

print("Setup complete. XGBoost available:", HAS_XGB)
```

## Load and inspect

```python
df = pd.read_csv("provider_features.csv")
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
print("\nFirst few rows:")
df.head()
```

## Clean data (convert percentages and types)

```python
# Strip any whitespace from column names
df.columns = df.columns.str.strip()

# Convert percentage columns that came in as strings (e.g. "45.5%")
def to_float(val):
    if pd.isna(val):
        return np.nan
    if isinstance(val, str):
        return float(val.replace('%', '').replace(',', '').strip()) / (100 if '%' in val else 1)
    return float(val)

# Apply to all columns except identifier/label/dataset
feature_cols = [c for c in df.columns if c not in ['Provider', 'Dataset', 'PotentialFraud']]
for col in feature_cols:
    df[col] = df[col].apply(to_float)

# Convert PotentialFraud to numeric label (1 = Yes, 0 = No, NaN = Test/unlabeled)
df['label'] = df['PotentialFraud'].map({'Yes': 1, 'No': 0})

print("Total rows:", len(df))
print("Labeled rows (Train):", df['label'].notna().sum())
print("Unlabeled rows (Test):", df['label'].isna().sum())
print("Fraud rate in Train:", df['label'].mean())
```

## Split Train into model train/val

```python
# Use only labeled rows for model training
train_df = df[df['label'].notna()].copy()
test_df = df[df['label'].isna()].copy()  # Kaggle Test set (to be scored)

# Fill missing feature values with 0 (some measures return blank for tiny providers)
train_df[feature_cols] = train_df[feature_cols].fillna(0)
test_df[feature_cols] = test_df[feature_cols].fillna(0)

X = train_df[feature_cols]
y = train_df['label'].astype(int)

# Stratified 80/20 split — same fraud rate in both halves
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("Train size:", len(X_train), "Val size:", len(X_val))
print("Fraud rate train:", y_train.mean(), "val:", y_val.mean())
```

## Train the model

```python
if HAS_XGB:
    model = XGBClassifier(
        n_estimators=200,
        max_depth=4,
        learning_rate=0.1,
        scale_pos_weight=10,  # account for class imbalance
        random_state=42,
        eval_metric='auc'
    )
else:
    model = RandomForestClassifier(
        n_estimators=200,
        max_depth=8,
        class_weight='balanced',
        random_state=42,
        n_jobs=-1
    )

model.fit(X_train, y_train)
print("Model trained.")
```

## Evaluate

```python
y_pred = model.predict(X_val)
y_proba = model.predict_proba(X_val)[:, 1]

print("AUC:", roc_auc_score(y_val, y_proba))
print("\nClassification Report:")
print(classification_report(y_val, y_pred, target_names=['Not Fraud', 'Fraud']))
print("\nConfusion Matrix:")
print(confusion_matrix(y_val, y_pred))
```

## Feature importance

```python
importances = pd.DataFrame({
    'feature': feature_cols,
    'importance': model.feature_importances_
}).sort_values('importance', ascending=False)

print(importances)

plt.figure(figsize=(10, 6))
plt.barh(importances['feature'], importances['importance'])
plt.gca().invert_yaxis()
plt.title('Feature Importance for Fraud Prediction')
plt.xlabel('Importance')
plt.tight_layout()
plt.show()
```

## Score ALL providers (train + test)

```python
# Score every row
df[feature_cols] = df[feature_cols].fillna(0)
df['fraud_probability'] = model.predict_proba(df[feature_cols])[:, 1]

# Bucket into risk bands
def risk_band(p):
    if p >= 0.7:
        return '1-High'
    elif p >= 0.3:
        return '2-Medium'
    else:
        return '3-Low'

df['ml_risk_band'] = df['fraud_probability'].apply(risk_band)

# Show summary
print("Provider count by ML risk band:")
print(df['ml_risk_band'].value_counts())
print("\nFraud probability distribution:")
print(df['fraud_probability'].describe())
```
